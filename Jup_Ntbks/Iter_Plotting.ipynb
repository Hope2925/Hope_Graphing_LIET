#### Plot the fit time distributions for each run with and without mLa (color code by whether or not with mLa)


```python
## violin plot with each column being the number of iterations for all genes
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
ax.set_title('Violin plot comparing fit time distributions for all genes')
ax.set_ylabel('Fit time (min)')
data = [iter1.df['fit_time_min'], iter3.df['fit_time_min'], iter5.df['fit_time_min'], iter1o.df['fit_time_min'], iter3o.df['fit_time_min'], iter5o.df['fit_time_min']]
ax.violinplot(data, showmeans=True)

def set_axis_style(ax, labels):
    ax.xaxis.set_tick_params(direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks(np.arange(1, len(labels) + 1), labels=labels)
    ax.set_xlim(0.25, len(labels) + 0.75)
    ax.set_xlabel('Sample name')

set_axis_style(ax, labels=['10 w/','30 w/', '50 w/', '10 w/o', '30 w/o', '50 w/o'])
plt.show()
```

```python
title = 'Fit time distributions for all genes between different # iterations, with or without antisense parameters'
y_label = 'Fit time (min)'
x_label = "Number of iterations (1000s) with or without antisense parameters"
labels=['10 w/','30 w/', '50 w/', '10 w/o', '30 w/o', '50 w/o']
meanpointprops = dict(marker='D', markeredgecolor='black', markerfacecolor='firebrick')
data = [iter1.df['fit_time_min'], iter3.df['fit_time_min'], iter5.df['fit_time_min'], iter1o.df['fit_time_min'], iter3o.df['fit_time_min'], iter5o.df['fit_time_min']]
pf.box_plotter(data, labels=labels, title=title, y_label=y_label, x_label=x_label, meanpointprops=meanpointprops)


datadict = {"1 w/": iter1.df['fit_time_min'], "3 w/": iter3.df['fit_time_min'], "5 w/":iter5.df['fit_time_min']}
title = "Fit times between different # iterations, with or without antisense parameters"
y_label = "Fit time (min)"
x_label = "Number of iterations (1000s) with or without antisense parameters"
pf.box_plotter_all(datadict, title=title, y_label=y_label, x_label=x_label showmeans=False)
```
